```python
import numpy as np
import pandas as pd

np.random.seed(42)
m = 100
X1 = 6 * np.random.rand(m, 1) - 3
X2 = 3.2 * np.random.rand(m, 1) - 1

y = 1.3 * X1 ** 2 + 0.5 * X ** 2 + 2 * X1 * X2 + np.random.randn(m , 1)

df = pd.DataFrame({
    'x1': X1.ravel(),
    'x2': X2.ravel(),
    'y': y.ravel()
})
df.head()
```

```python
from sklearn.metrics import r2_score


x = df.drop(columns='y')
y = df['y']

reg_linear = LinearRegression()
reg_linear.fit(x,y)

y_pred = reg_linear.predict(x)
r2_score(y, y_pred)
```

```python
poly_feat = PolynomialFeatures(degree=2, include_bias=False)

x_poly = pd.DataFrame(poly_feat.fit_transform(x), columns=['x1', 'x2', 'x1_quad', 'x1x2', 'x2_quad'])
x_poly.head()
```

```python
linear_reg = LinearRegression()
linear_reg.fit(x_poly,y)
```

```python
r2_score(y, linear_reg.predict(x_poly))
```
